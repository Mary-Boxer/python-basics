```python
import pandas as pd
import matplotlib.pyplot as plt

# Данные
data = {
    "product": ["Ball", "Ring", "Jump rope", "Ball", "Ring", "Ball", "Jump rope", "Ring"],
    "store": ["A", "A", "B", "B", "C", "C", "A", "B"],
    "quantity": [9, 4, 7, 3, 5, 2, 6, 8],
    "price": [500, 550, 400, 500, 530, 600, 450, 560]
}
df = pd.DataFrame(data)

# Считаем общее количество по каждому продукту
total_quantity = df.groupby("product")["quantity"].sum().sort_values(ascending=False)

# Берём топ-3
top3 = total_quantity.head(3)

# Строим график
plt.figure(figsize=(6,4))
top3.plot(kind="bar", color="skyblue", edgecolor="black")

plt.title("Top-3 products by total sales")
plt.ylabel("Total quantity sold")
plt.xlabel("Product")
plt.xticks(rotation=0)

plt.show()
```
